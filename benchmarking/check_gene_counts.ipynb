```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
gene_counts = {}
with open('../data/l1000_cp.gmt') as f:
    lines = f.readlines()
    for l in tqdm(lines):
        l = l.strip()
        for g in l.split('\t')[2:]:
            if g in gene_counts:
                gene_counts[g] += 1
            else:
                gene_counts[g] = 1
                

with open('../data/l1000_xpr.gmt') as f:
    lines = f.readlines()
    for l in tqdm(lines):
        l = l.strip()
        for g in l.split('\t')[2:]:
            if g in gene_counts:
                gene_counts[g] += 1
            else:
                gene_counts[g] = 1
    
```

```python
gene_count_df = pd.DataFrame(gene_counts.items(), columns=['gene', 'count'])
```

```python
n = 1436110 + 281890
gene_count_df['pct'] = (gene_count_df['count'] / n) * 100
```

```python
len(gene_count_df)
```

```python
gene_count_df.hist(column='pct', bins=50)
plt.ylabel('genes')
plt.xlabel('% sets containing gene')
plt.title('')
```

```python
gene_count_df.sort_values('pct', ascending=False).head(30).set_index('gene', drop=True)
```
